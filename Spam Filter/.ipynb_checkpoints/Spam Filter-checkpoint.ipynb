```python
# funckje

def counts_pct(category,data):
    '''Funkcja, która zwraca dla tanej kategori liczbe wartości, oraz wartość procentową'''
    base = data[category].value_counts()
    pct = base/data[category].shape[0] * 100
    joined = pd.concat([base,pct],axis=1)
    joined.columns = [str(category),str(category) + ' %']
    return joined

def annotate(ax,s="%.0f",suffix=''):
    for p in ax.patches:
        ax.annotate(s % p.get_height() + suffix, 
                 (p.get_x() + p.get_width() / 2,
                  p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
        
def despine(rot=0):
    sns.despine(left=True)
    plt.xticks(rotation = rot);
    plt.xlabel('');
    plt.ylabel('');

    
    
clr = ['#b3b7b8','#c55572']
```

### Przetwarzanie Języka Naturalnego (NLP)

Przetwarzanie języka naturalnego (NLP) to jeden z obszarów sztucznej inteligencji, który zajmuje się rozumieniem i przetwarzaniem języka ludzkiego. W świetle nowych postępów w uczeniu maszynowym wiele organizacji zaczęło stosować przetwarzanie języka naturalnego do tłumaczeń, chatbotów i filtrowania kandydatów.

Algorytmy uczenia maszynowego nie mogą bezpośrednio pracować z surowym tekstem. Zamiast tego - tekst musi zostać przekonwertowany na wektory liczb. W przetwarzaniu języka naturalnego powszechną techniką wyodrębniania elementów z tekstu jest umieszczanie wszystkich słów występujących w tekście w tzw "Bag Of Words"
Zazwyczaj pracując z takim "Bag of Words" możemy łatwo pozbyć się duplikatów z naszych danych za pomocą metody tekstowej Set w pythonie.

### TF-IDF

__TF-IDF__ jest to jedna z metod w NLP, która polega na ocenie względnej ważności słów w zbiorze danych. 
Jest to iloczyn częstotliwości wyrażenia oraz odwrotnej częśtości w dokumentach. Dzięki temu otrzymujemy wagę danego słowa/frazy w zbiorze badanych dokumentów


__TF__ (Term Frequency) - mówi o tym jak często dane słowo występuje w konkretnym dokumencie. Jego zakres obejmuje wyłącznie jeden wiersz (dokument) Wartość ta jest wprost proporcjonalna do tego jak często dany wyraz występuje.
TF obliczamy w następujący sposób: dzielimy liczbę wystąpień frazy w (wierszu) dokumencie przez liczbę znajdujących się w nim wszystkich słów. 

__IDF__ (Inverse Document Frequency) mówi o tym jak często dane słowo występuje we wszystkich wierszach (dokumentach) zbiorze badanych dokumentów. Jest to odwrotna częstotliwość wyrażenia, co znaczy, że im częściej dany wyraz pojawia się zbiorze danych, tym wynik IDF będzie niższy.


Algorytm stosowany jako metoda oceny relewantności dokumentu w wyszukiwarkach internetowych, kolejnym zastosowaniem jest ocena podobieństwa dokumentów w systemach grupowania wyników oraz systemach typu antyplagiat.




Kiedy mamy do czynienia z problemem tekstowym w przetwarzaniu języka naturalnego (NLP) , proces usuwania 'Stop-Words' jest jednym z ważniejszyc kroków, który powoduje że mamy większą dokładność naszego modelu. Stop-Words oznaczają bardzo popularne słowa w danym języku (np. A, an, po angielsku). Słowe te zazwyczaj nie mają zbytniego impaktu w przy budowaniu modeli w zakresie NLP, szczególnie przy problemach tj. analiza semantyczna, klasyfikacja.


Poniżej mamy listę stopwords z jezyka angielskiego:  

```python
print(set(stopwords.words('english')))
```

```python
def text_preprocess(text):
    '''Funkcja ma na celu usunięcie znaków tj. [!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~]
    Usunięcia 'Stop Words' w NLP (Natural Language Processing) - Sa to słowa, określane jako bezużyteczne
    '''
    delete_punctuation = text.translate(str.maketrans('', '', string.punctuation))
    text = [word for word in delete_punctuation.split() if word.lower() not in  stopwords.words('english')]
    return " ".join(text)
```

```python
data['text_processed'] = data['text'].apply(lambda x: text_preprocess(x))
```

```python
vectorizer = TfidfVectorizer("english")
```

```python
text_vectorized = vectorizer.fit_transform(data['text_processed'])
```

```python
text_vectorized
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(text_vectorized, data['spam'], test_size=0.3, random_state=42)
```

### Model SVC - Support Vector Classifie

Model ten jest szeroko stosowany w celach klasyfikacji. SVC to klasyfikator formalnie zdefiniowany przez oddzielającą hiperpłaszczyznę. Innymi słowy, biorąc pod uwagę oznaczone dane treningowe (uczenie nadzorowane), algorytm generuje optymalną hiperpłaszczyznę, która kategoryzuje nowe przykłady. W dwóch wymiarach przestrzennych hiperpłaszczyzna jest linią dzielącą płaszczyznę na dwie części, gdzie klasy leżą po obu stronach. Aby oddzielić dwie klasy punktów danych, można wybrać wiele możliwych hiperpłaszczyzn. Naszym celem jest znalezienie płaszczyzny, która ma maksymalny margines, tj. Maksymalną odległość między punktami danych obu klas. 

```python
from sklearn.svm import SVC
svc = SVC(kernel='sigmoid',gamma=1)
svc.fit(X_train, Y_train)
predict_svc = svc.predict(X_test)
cnf_matrix_svc = confusion_matrix(Y_test, predict_svc)
```

```python
print('Skuteczność dla modelu SVC: ', round(accuracy_score(Y_test,predict_svc),4)*100 , ' %')
```

```python
cnf_matrix_svc
```

```python
fig, ax = plot_confusion_matrix(conf_mat=cnf_matrix_svc,
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True)
```

```python
print(classification_report(Y_test, predict_svc))
```

### Random Forest Classifier

RandomForestClassifier - Jednen z najleszych i najczęściej stosowanych modeli do klasyfikacji.
Losowy las składa się z dużej liczby pojedynczych drzew decyzyjnych, które działają jako zespół. Każde pojedyncze drzewo w losowym lesie wyrzuca prognozę klasy, a klasa z największą liczbą głosów staje się prognozą naszego modelu. Kluczem RandomForestClassifier jest niska korelacja między drzewami. Nieskorelowane modele mogą wytwarzać prognozy zestawów, które są dokładniejsze niż jakiekolwiek z indywidualnych prognoz. Powodem tego cudownego efektu jest to, że drzewa chronią się nawzajem przed swoimi indywidualnymi błędami (o ile nie wszystkie ciągle błądzą w tym samym kierunku). Podczas gdy niektóre drzewa mogą się mylić, wiele innych drzew będzie miało rację, więc jako grupa drzewa mogą poruszać się we właściwym kierunku.

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=50, n_jobs=-1)
rf_model = rf.fit(X_train, Y_train)
rf_prediction=rf_model.predict(X_test)
cnf_matrix_rfc = confusion_matrix(Y_test, rf_prediction)
print('Skuteczność dla modelu RandomForestClassifier: ', round(accuracy_score(Y_test,rf_prediction),4)*100 , ' %')
```

```python
fig, ax = plot_confusion_matrix(conf_mat=cnf_matrix_rfc,
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True,cmap='Oranges')
```

```python
print(classification_report(Y_test, rf_prediction))
```

### Naiwny klasyfikator bayesowski 

prosty klasyfikator probabilistyczny. Naiwne klasyfikatory bayesowskie są oparte na założeniu o wzajemnej niezależności predyktorów (zmiennych niezależnych). Często nie mają one żadnego związku z rzeczywistością i właśnie z tego powodu nazywa się je naiwnymi. Bardziej opisowe jest określenie – „model cech niezależnych”. Pomimo ich naiwnego projektowania i bardzo uproszczonych założeń, w wielu rzeczywistych sytuacjach naiwne klasyfikatory Bayesa często pracują dużo lepiej, niż można było tego oczekiwać.


Nazywa się to naiwny klasyfikator bayesowski , ponieważ obliczenia prawdopodobieństw dla każdej hipotezy są uproszczone, tak aby wogóle były wykonalne. Zamiast próbować obliczyć wartości każdej wartości atrybutu P (d1, d2, d3 | h), zakłada się, że są one warunkowo niezależne z uwagi na wartość docelową i obliczane jako P (d1 | h) * P (d2 | H) i tak dalej. 
Jest to bardzo naiwne założenie, które jest mało prawdopodobne w rzeczywistych danych, tzn. ze atrybuty nie wchodzą w interakcje. Niemniej jednak podejście to działa zaskakująco dobrze z danymi, w przypadku których to założenie nie ma zastosowania.


```python
from sklearn.naive_bayes import MultinomialNB
multi_nb = MultinomialNB(alpha=0.2)
nb_model = multi_nb.fit(X_train, Y_train)
nb_prediction=nb_model.predict(X_test)
cnf_matrix_nb = confusion_matrix(Y_test, nb_prediction)
print('Skuteczność dla modelu Naive Bayes: ', round(accuracy_score(Y_test, nb_prediction),4)*100 , ' %')
```

```python
fig, ax = plot_confusion_matrix(conf_mat=cnf_matrix_nb,
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True,cmap='Reds')
```

```python
print(classification_report(Y_test, nb_prediction))
```

### DecisionTreeClassifier
Drzewo decyzyjne to prosta metoda do klasyfikowania przykładów. Jest to nadzorowane uczenie maszynowe, w którym dane są stale dzielone zgodnie z określonym parametrem. Drzewa decyzyjne są popularną nadzorowaną metodą uczenia się. Zalety drzew decyzyjnych obejmują to, że można ich używać zarówno do regresji, jak i klasyfikacji, są łatwe do interpretacji i nie wymagają skalowania funkcji.

```python
fig, ax = plot_confusion_matrix(conf_mat=cnf_matrix_dt,
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True,cmap='cividis_r')
```

```python
print(classification_report(Y_test, dt_prediction))
```

### K najbliższych sąsiadów - KNeighborsClassifier

Metoda K Najbliższych Sąsiadów (k-Nearest Neighbors) należy do grupy algorytmów leniwych (lazy algorithms), czyli takich, które nie tworzą wewnętrznej reprezentacji danych uczących, lecz szukają rozwiązania dopiero w momencie pojawienia się wzorca testującego, np. do klasyfikacji. Przechowuje wszystkie wzorce uczące, względem których wyznacza odległość wzorca testowego. Algorytm regresji nieparametrycznej używany w statystyce do prognozowania wartości pewnej zmiennej losowej). Leniwe uczenie oznacza, że ​​algorytm nie dokonuje uogólnień. Oznacza to, że korzystanie z tej metody wymaga niewielkiego szkolenia. Z tego powodu wszystkie dane szkoleniowe są również wykorzystywane do testowania podczas korzystania z KNN.

```python
from sklearn.neighbors import KNeighborsClassifier

knc = KNeighborsClassifier()
knc_model = knc.fit(X_train, Y_train)
knc_prediction=knc_model.predict(X_test)
cnf_matrix_knc = confusion_matrix(Y_test, knc_prediction)
print('Skuteczność dla modelu KNeighbors: ', round(accuracy_score(Y_test,knc_prediction),4)*100 , ' %')
```

```python
fig, ax = plot_confusion_matrix(conf_mat=cnf_matrix_knc,
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True,cmap='viridis_r')
```

```python
print(classification_report(Y_test, knc_prediction))
```
